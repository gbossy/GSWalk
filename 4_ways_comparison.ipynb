```python
#Trying to see how to minimize the effect of the norm on coloring order
n=50
mu=1
for i in range(repeat):
    v_normalized=[]
    v_add1=[]
    v_addn=[]
    for i in range(n):
        v_normalized.append(normalize(v[i]))
        counterweight=np.sqrt(mu-norm(v[i]))
        v_add1.append(np.append(v[i],counterweight))
        added=np.zeros(n)
        added[i]=counterweight
        v_addn.append(np.concatenate((v[i],added)))
    B=np.transpose(np.vstack(tuple([e for e in v])))
    for vecs in [v,v_normalized,v_add1,v_addn]:
        gsw_xs=[]
        when_colored=[]
        for p in range(n):#[195,196,197,198,199]:
            #print(f'\n Try #{p}')
            res,order=gram_schmidt_walk(v,np.zeros(n),order=True,smallest_delta=False,mode='random',pivot=p)
            gsw_xs.append(res)
            switched_list=[(i,order[i]) for i in range(len(order))]
            switched_list.sort(key=lambda x:x[1])
            when_col=[x[0] for x in switched_list]
            when_colored.append(when_col)
        print(f'\n Final breakdown')
        #Average output norm, average moment of coloring variance, average group balance,
        avg_output_norm=norm(np.matmul(B,average(gsw_xs)))
        print(f'Average output norm with original vectors: {avg_output_norm}')
        avg_var=average(np.var(np.array(when_colored),axis=0))
        print(f'Average variance of coloring step: {avg_var}')
        print(f'Average coloring balance:{average([abs(sum(x)) for x in gsw_xs])}')
    print('\n--------------------------')
```
