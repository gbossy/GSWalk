```python
foldername='I_and_D_instead_of_D_inv_example/'
def run_experiment(n,t=0):
    v=sample_from_ball(n,d=2)
    B=np.transpose(np.vstack(tuple([e for e in v])))
    a,b=-np.ones(n),np.ones(n)
    gsw_xs=[]
    gsw_xs_with_I=[]
    gsw_xs_with_D=[]
    for i in range(n):
        gsw_xs.append(gram_schmidt_walk(v,np.zeros(n)))
        gsw_xs_with_I.append(gram_schmidt_walk(v,np.zeros(n),i_instead_of_d_inv=True))
        gsw_xs_with_D.append(gram_schmidt_walk(v,np.zeros(n),d_instead_of_d_inv=True))
    plt.plot([np.matmul(B,x)[0] for x in gsw_xs],[np.matmul(B,x)[1] for x in gsw_xs],'o',label='GSW')
    plt.plot([np.matmul(B,x)[0] for x in gsw_xs_with_I],[np.matmul(B,x)[1] for x in gsw_xs_with_I],'o',label='GSW $\cdot D$')
    plt.plot([np.matmul(B,x)[0] for x in gsw_xs_with_D],[np.matmul(B,x)[1] for x in gsw_xs_with_D],'o',label='GSW $\cdot D^2$')
    plt.legend()
    plt.legend(bbox_to_anchor=(1.35, 1))
    tikzplotlib.save(f"{foldername}3types_basis_{n}_{t}.tex")
    plt.show()
```

```python
for i in range(3):
    run_experiment(100,i)
```

```python
for i in range(10):
    run_experiment(200,i)
```
