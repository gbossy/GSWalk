```python
n=200
bigs=10
v=sample_from_ball(200,d=100)
for i in range(bigs):
    v[i]=10*normalize(v[i])
gsw_xs=[]
avg_time_step_of_coloring_for_smaller_dimensioned=[]
for i in range(100):
    print(f'\n Try #{i}')
    res,order=gram_schmidt_walk(v,np.zeros(n),order=True,smallest_delta=False)
    gsw_xs.append(res)
    avg_step_struct=average([j for j in range(len(order)) if order[j]<bigs])
    print(f'Average time step of coloring of big elements: {avg_step_struct}')
    avg_time_step_of_coloring_for_smaller_dimensioned.append(avg_step_struct)
print(f'\n Final breakdown')
print(f'Average of Average time step of coloring of long elements: {average(avg_time_step_of_coloring_for_smaller_dimensioned)}')
```
