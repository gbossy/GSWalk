```python
#How does translation of the input affect the distribution of 1 and -1's? Here no translation
```

```python
def run_experiment(v,det=False):
    gsw_xs=[]
    balances=[]
    for i in range(100):
        print(f'\n Try #{i}')
        res,order=gram_schmidt_walk(v,np.zeros(n),order=True,smallest_delta=det)
        gsw_xs.append(res)
        bal=abs(sum(res))
        print(f'Balance of the coloring: {bal}')
        balances.append(bal)

    print(f'\n Final breakdown')
    print(f'Average of balances of the colorings:{average(balances)}')
```

```python
#GSW with no translation
n=200
v=sample_from_ball(n,d=n)
run_experiment(v)
```

```python
#Deterministic GSW with no translation
n=200
v=sample_from_ball(n,d=n)
run_experiment(v,det=True)
```

```python
#GSW + small Translation
n=200
v=sample_from_ball(n,d=n)
trans=sample_from_ball(1,d=n)
for i in range(len(v)):
    v[i]+=normalize(trans[0])
run_experiment(v)
```

```python
#Deterministic GSW + small Translation
n=200
v=sample_from_ball(n,d=n)
trans=sample_from_ball(1,d=n)
for i in range(len(v)):
    v[i]+=normalize(trans[0])
run_experiment(v,det=True)
```
