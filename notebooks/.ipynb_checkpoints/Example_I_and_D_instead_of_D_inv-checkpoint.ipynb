```python
foldername='Example_basis/'
def run_experiment(n,t=0):
    x=np.random.uniform(size=(n,n))
    basis_1=[normalize(x[i,:]) for i in range(x.shape[0])]
    x=np.random.uniform(size=(n,n))
    basis_2=gram_schmidt_columns(x)
    v=sample_from_ball(n,d=2)
    B=np.transpose(np.vstack(tuple([e for e in v])))
    a,b=-np.ones(n),np.ones(n)
    gsw_xs=[]
    gsw_basis_xs_1=[]
    gsw_basis_xs_2=[]

    for i in range(n):
        gsw_xs.append(gram_schmidt_walk(v,np.zeros(n)))
        gsw_basis_xs_1.append(gram_schmidt_walk(v,np.zeros(n),a,b,False,False,False,basis_1))
        gsw_basis_xs_2.append(gram_schmidt_walk(v,np.zeros(n),a,b,False,False,False,basis_2))
    plt.plot([np.matmul(B,x)[0] for x in gsw_xs],[np.matmul(B,x)[1] for x in gsw_xs],'o',label='GSW')
    plt.plot([np.matmul(B,x)[0] for x in gsw_basis_xs_1],[np.matmul(B,x)[1] for x in gsw_basis_xs_1],'o',label='GSW B1')
    plt.plot([np.matmul(B,x)[0] for x in gsw_basis_xs_2],[np.matmul(B,x)[1] for x in gsw_basis_xs_2],'o',label='GSW B2 ')
    plt.legend()
    plt.legend(bbox_to_anchor=(1.35, 1))
    tikzplotlib.save(f"{foldername}3types_basis_{n}_{t}.tex")
    plt.show()
```

```python
for i in range(3):
    run_experiment(100,i)
```

```python
for i in range(10):
    run_experiment(200,i)
```
