```python
#Trying to see what's colored early in 2 dimension
n=50
#v=sample_from_sphere(n,d=2)
v=sample_from_ball(n,d=2)
for vec in v:
    vec=normalize(vec)
#print(f'v:{v}')
plt.plot([e[0] for e in v],[e[1] for e in v],'o',label='Input vector')
B=np.transpose(np.vstack(tuple([e for e in v])))
gsw_xs=[]
when_colored=[]
for i in range(100):
    #print(i)
    x,order=gram_schmidt_walk(v,np.zeros(n),plot=False,order=True,debug=False,no_matrix_mult=False)
    gsw_xs.append(x)
    switched_list=[(i,order[i]) for i in range(len(order))]
    switched_list.sort(key=lambda x:x[1])
    when_col=np.array([x[0] for x in switched_list])
    when_colored.append(when_col)
plt.plot([np.matmul(B,x)[0] for x in gsw_xs],[np.matmul(B,x)[1] for x in gsw_xs],'o',label='Output sum of groups with GSW')
ACT=average(when_colored)
print(ACT)
for i in range(len(v)):
    if ACT[i]<n/4 or ACT[i]>3*n/4:
        plt.plot(v[i][0],v[i][1],'o',label=f'ACT: {ACT[i]}')
plt.legend()
plt.legend(bbox_to_anchor=(1.35, 1))
```

```python
#Trying to see what's colored early in 2 dimension
n=50
#v=sample_from_sphere(n,d=2)
v=sample_from_ball(n,d=2)
for i in range(len(v)):
    v[i]=normalize(v[i])
print(v)
#print(f'v:{v}')
plt.plot([e[0] for e in v],[e[1] for e in v],'o',label='other input vectors')
B=np.transpose(np.vstack(tuple([e for e in v])))
gsw_xs=[]
when_colored=[]
for i in range(100):
    #print(i)
    x,order=gram_schmidt_walk(v,np.zeros(n),plot=False,order=True,debug=False,no_matrix_mult=False,mode='max_coloring')
    gsw_xs.append(x)
    switched_list=[(i,order[i]) for i in range(len(order))]
    switched_list.sort(key=lambda x:x[1])
    when_col=np.array([x[0] for x in switched_list])
    when_colored.append(when_col)
plt.plot([np.matmul(B,x)[0] for x in gsw_xs],[np.matmul(B,x)[1] for x in gsw_xs],'o',label='Output sum of groups with GSW')
ACT=average(when_colored)
print(ACT)
early_vs=[v[i] for i in range(len(v)) if ACT[i]<n/4]
plt.plot([vec[0] for vec in early_vs],[vec[1] for vec in early_vs],'o',label=f'Early')
late_vs=[v[i] for i in range(len(v)) if ACT[i]>3*n/4]
plt.plot([vec[0] for vec in late_vs],[vec[1] for vec in late_vs],'o',label=f'Late')
plt.legend()
plt.legend(bbox_to_anchor=(1.35, 1))
```
