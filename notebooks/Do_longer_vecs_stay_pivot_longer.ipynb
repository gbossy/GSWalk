```python
def run_experiment(v,det=False):
    n=len(v)
    group_sep=int(n/2)
    gsw_xs=[]
    how_long={0:[],1:[]}
    for i in range(100):#range(int(n/2)):
        print(f'\n Try #{i}')
        res,pivots=gram_schmidt_walk(v,np.zeros(n),return_pivots=True,smallest_delta=det)
        gsw_xs.append(res)
        rep=1
        for j in range(n-1):
            if pivots[j]==pivots[j+1]:
                rep+=1
            else:
                if pivots[j]<n/2:
                    how_long[0].append(rep)
                else:
                    how_long[1].append(rep)
                rep=1
    print(f'\n Final breakdown')
    print(f'Average length of stay at pivot for group 0:{average(how_long[0])}')
    print(f'Average length of stay at pivot for group 1:{average(how_long[1])}')
```

```python
#Now with a group of big vectors and a group of small vectors and classical GSW
n=200
v=sample_from_ball(n,d=n)
for i in range(len(v)):
    v[i]=normalize(v[i])*(1 if i<n/2 else n)
run_experiment(v)
```

```python
#Now with a group of big vectors and a group of small vectors and det GSW
n=200
v=sample_from_ball(n,d=n)
for i in range(len(v)):
    v[i]=normalize(v[i])*(1 if i<n/2 else n)
run_experiment(v,det=True)
```
